4. Write a function to convert the first letter of each word in a string to uppercase and the rest lowercase without using title().

```python
def to_title(str):
    res=''
    capitalize=True
    for ch in str:
        code=ord(ch)
        if ch == " ":
            res+=ch
            capitalize=True
        else:
            if capitalize and 'a' <= ch <= 'z':
                res+= chr(code -32)
                capitalize=False
            elif not capitalize and 'A' <= ch <= 'Z':
                res+= chr(code +32)
            else:
                res+=ch
                capitalize=False
    return res
print(to_title("hello world! this is a title."))            

```

5. Write a function to capitalize the first character of a string and convert the rest into lowercase without using capitalize().

```python
def capitalize(str):
    res=''
    for i in range(len(str)):
        code=ord(str[i])
        if i==0 and 'a' <= str[i] <= 'z':
            res+= chr(code -32)
        elif i!=0 and 'A' <= str[i] <= 'Z':
            res+= chr(code +32)
        else:
            res+= str[i]
    return res
print(capitalize("hello World! this is a title."))
        
```
